# Milestone 1

```python
import pandas as pd
import numpy as np
```

```python
column_names=[
    'engine_id','cycle','op_setting_1','op_setting_2','op_setting_3',
    'sensor_1','sensor_2','sensor_3','sensor_4','sensor_5','sensor_6',
    'sensor_7','sensor_8','sensor_9','sensor_10','sensor_11','sensor_12',
    'sensor_13','sensor_14','sensor_15','sensor_16','sensor_17',
    'sensor_18','sensor_19','sensor_20','sensor_21'
]
file_path='train_FD001.txt'
```

```python
train_df=pd.read_csv(file_path,sep='\s+',header=None,names=column_names)
print("Data from train_FD001.txt loaded successfully.")
```

```python
max_cycles_df = train_df.groupby('engine_id')['cycle'].max().reset_index()
max_cycles_df.columns=['engine_id','max_cycle']
```

```python
train_df = pd.merge(train_df, max_cycles_df, on='engine_id', how='left')
train_df['RUL'] = train_df['max_cycle'] - train_df['cycle']
train_df.drop('max_cycle', axis=1, inplace=True)
```

```python
print("DataFrame with the calculated RUL column:")
train_df.head()
```

```python
engine_id_to_check = 1
single_engine_df = train_df[train_df['engine_id'] == engine_id_to_check]

print(f"--- Data for Engine ID: {engine_id_to_check} (First 5 cycles) ---")
print(single_engine_df.head())

print(f"\n--- Data for Engine ID: {engine_id_to_check} (Last 5 cycles) ---")
print(single_engine_df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
engines_to_plot = [1, 5, 10, 15]
sensors_to_plot = ['sensor_2', 'sensor_3', 'sensor_4', 'sensor_7', 'sensor_12']

subset_df = train_df[train_df['engine_id'].isin(engines_to_plot)]

for sensor in sensors_to_plot:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=subset_df, x='cycle', y=sensor, hue='engine_id', palette='viridis')
    plt.title(f'Sensor: {sensor} | Trend over Lifecycle for Sample Engines')
    plt.xlabel('Cycle')
    plt.ylabel('Sensor Reading')
    plt.legend(title='Engine ID')
    plt.show()
```

```python
data_std = train_df.std()
print("--- Standard Deviation of Each Column ---")
print(data_std)

constant_columns = data_std[data_std == 0].index.tolist()
print(f"\nConstant columns identified: {constant_columns}")

train_df.drop(columns=constant_columns, inplace=True)

print("\nConstant columns have been dropped from the DataFrame.")
print(f"Remaining columns: {train_df.columns.tolist()}")
```

```python
def generate_sequences(df, sequence_length, feature_cols, target_col):
    sequences = []
    labels = []

    for engine_id in df['engine_id'].unique():
        engine_df = df[df['engine_id'] == engine_id]

        sensor_data = engine_df[feature_cols].values
        rul_data = engine_df[target_col].values

        for i in range(len(engine_df) - sequence_length + 1):
            sequences.append(sensor_data[i:i+sequence_length])
            labels.append(rul_data[i+sequence_length-1])
            
    return np.array(sequences), np.array(labels)

sequence_length = 50 

feature_cols = [col for col in train_df.columns if 'sensor' in col]

X_train, y_train = generate_sequences(train_df, sequence_length, feature_cols, 'RUL')

print("--- Generated Sequences ---")
print(f"Shape of training sequences (X_train): {X_train.shape}")
print(f"Shape of training labels (y_train): {y_train.shape}")
```

# Milestone 2

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(
    units=50, 
    return_sequences=True, 
    input_shape=(X_train.shape[1], X_train.shape[2])
))
model.add(Dropout(0.2))
model.add(LSTM(
    units=50, 
    return_sequences=False
))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
print("Model compiled successfully and is ready for training.")
```

```python
epochs=10
batch_size=32
```

```python
history = model.fit(
    X_train, 
    y_train, 
    epochs=epochs, 
    batch_size=batch_size, 
    validation_split=0.2,
    verbose=1
)
print("\nModel training complete.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')

plt.title('Model Training and Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.show()
```
